## Comparing the dataset with my generated hypothesis

We can draw the following data points from the dataset according to my hypothesis:

#### 1. Geographic location: 
The geographic location accroding to the hypothesis can be taken up with the Historic town of Ames, Iowa

#### 2. Neighbourhood:
The nieghbourhood gives us details like the location in the city/town. The <i>neighbourhood</i> factor in the dataset provides us with the information we aim to focus on and <i>Zoning classification</i> adds more information to the data of neighbourhood. 

#### 3. Number of bedrooms, bathrooms and Balcony:
The number of bedrooms, bathrooms and balcony can be determined in the dataset from the data points <i>Full Bath, Bedrooms, Halfbath</i>. More the number of rooms with the increased surface area, more the sale price of the house.

#### 4. The area of the house
Area of the house includes the area of all the floors, garage and backyard or frontyard. The data points determining these values are <i>Garage Area, Firstfloor square feet, secondfloor square feet, wood porch area, enclosed porch area, open porch area, screen porch, pool area</i>

#### 5. Number of floors
Number of floors allows privacy and hence is being preffered

#### 6. Parking and basement area
The area of parking and basement area have been incuded in the total area

#### 7. Furnished and unfurnished
This feature determines whether the bedrooms are furnished with beds and neccessary furniture or not. The dataset does not describe this feature

#### 8. Year of construction
The data point <i>YearBuilt and Garage YearBuilt</i> determines the year in which the house was built. This overall affects the sales price of the house

#### 9. Utilities
Utilities in the dataset includes points: <i>Heating system, electrical system, central air, number of fireplace and utilities</i>
