## Abstract

House prices continue to grow at a rapid pace every year. There has been an increasing demand of the first time house buyers. There is a simple rule, the prices are expected to increase with the increasing demand for any product in the market. There are economic conditions of the country, state and city that also influences the prices of the house. Other than the demand-supply cycle and the economic conditions, there are various factors that influence the prices of the house. By leveraging the existing dataset of house purchases made in Ames, Iowa between the year 2006 and 2010, I focus here to build a statistical model to predict house prices in Ames.

The purpose of predictive modeling research was to identify the factors that demonstrated significant statistical correlations with House Prices, in other words, could I identify the factors that influences the house prices the most? I seek to develop practical predictive models capable of predicting prices using a subset of the significant factors. In an effort to develop a more scalable model, I also recognized some factors which could linearly affect the house prices. This research was driven by the long-term objective to facilitate and strengthen the understanding between the factors of the house and their sale prices. 
This research aimed to answer three primary questions: 
1. Which factors are the most influential factors for Sale Price of house in Ames, Iowa? 
2. How can marketers utilize the model to understand the past, current and future factors affecting the house prices with a goal of improving housing market in Ames, Iowa?

The dataset used in this project was collected and made available to public by <b>Dean De Cock</b>.

>#### Details of the population selected for building predictive model

> The dataset has information of house details and prices of sales of houses from year the 2006 to year the 2010 in Ames, Iowa in The United States of America. The house prices vary for same type of house in each city or town and hence knowing more about Ames was very important.[2] Ames is a city located in the central part of Story County, Iowa, with an area of 23.5 sq miles. The historical city had an population of 59,000 in 2010 with 84.3% white population. Ames is home of Iowa State University of Science and Technology, a public land-grant and space-grant research university which employees 15,500 people in the city. It has 37% lower crime rate thaan the national average and a pleasant weather through the year. The city has lower median earning compared to the national average and on the same hand, 16% lower unemployement rate than the national average.

>### -  __Model 3: Decision Tree Regression Model __<br>
>[4]Decision tree builds regression or classification models in the form of a tree structure. It brakes down a dataset into smaller and smaller subsets while at the same time an associated decision tree is incrementally developed. The final result is a tree with decision nodes and leaf nodes. A decision node (e.g., Outlook) has two or more branches (e.g., Sunny, Overcast and Rainy), each representing values for the attribute tested. Leaf node (e.g., Hours Played) represents a decision on the numerical target. The topmost decision node in a tree which corresponds to the best predictor called root node. Decision trees can handle both categorical and numerical data.<br>

### Conclusion

<p>The prediction models built in this project utilised certain factors of the house and the location. It can be said from the models that sale prices are highly influenced by the area of the house, the basement area and the number of cars that can be accomodated in the housing. These proved to be most influencial as the model attained 67% accuracy right with these three factors. The other factors like neighborhood, house and building type followed in the influencial factors list.The best model observed 81% accuracy on cross validation score which according to me has a scope of improvement.</p>

<p> A software version of the project can help the marketers predict the prices just on few selected factors used here. This project does not have modern data, which means the changes in neighborhood, building and house type preference over the period of 2010 - 2018 could have changed and can give us lower accuracy. If Ames had observed great economic growth over the period the model would need more data to learn and improve the model for better predictions.  </p>
